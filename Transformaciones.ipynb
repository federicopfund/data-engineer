```
Categoria.show()
FactMine.show()
Mine.show()
Producto.show()
Subcategoria.show()
Sucursales.show()
VentasInternet.show()
```

```
# TRANSFORMACION 1: En la tabla “Categoria” renombrar el campo Categoria a Nombre_Categoria.
transformacion1 = Categoria.withColumnRenamed("Categoria", "Nombre_Categoria")
transformacion1.show()
```

```
# TRANSFORMACION 2: En la tabla “FactMine” obtener el total de la columna: TotalOreMined.
transformacion2 = FactMine.agg(F.sum("TotalOreMined").alias("Suma_TotalOreMined"))
transformacion2.display()
```

```
# TRANSFORMACION 3: En la tabla “Mine” seleccionar únicamente las columnas: Country, FirstName,LastName y Age.

transformacion3 = Mine.select("Country","FirstName","LastName","Age")
transformacion3.display()
```

```
# TRANSFORMACION 4: En la tabla “Mine” obtener el total de la columna: TotalWasted agrupado por el atributo Country.

transformacion4 = Mine.groupBy("Country").agg(F.sum("TotalWasted").alias("Suma_TotalWasted"))
transformacion4.display()
```

```
# TRANSFORMACION 5: En la tabla “Producto” realizar un conteo de la cantidad de productos disponibles utilizando el campo: Cod_Producto.

transformacion5 = Producto.agg(F.count("Cod_Producto").alias("Cantidad_CodProducto"))
transformacion5 = transformacion5.withColumn("Cantidad_CodProducto" ,F.col("Cantidad_CodProducto").cast(IntegerType()))
transformacion5.display()
```

```
# TRANSFORMACION 6: En la tabla “Producto” mostrar de forma descendiente la información ordenada por el campo: Cod_Producto.

transformacion6 = Producto.sort(F.col("Cod_Producto").desc())
transformacion6.display()
```

```
# TRANSFORMACION 7: En la tabla “Subcategoria” realizar un filtro por el Cod_Categoria = 3.

transformacion7 = Subcategoria.filter(Subcategoria.Cod_Categoria == 3)
transformacion7.display()
```

```
# TRANSFORMACION 8: En la tabla “VentasInternet” generar una columna llamada Ingresos Netos, que se obtenga de multiplicar los atributos: (Cantidad * PrecioUnitario) – CostoUnitario.

transformacion8 = VentasInternet.withColumn("Ingresos_Netos", F.round(F.col("Cantidad")*F.col("PrecioUnitario")-F.col("CostoUnitario"),4).cast(DecimalType(10,4)))

transformacion8.display()
```

```
# TRANSFORMACION 9: En la tabla “VentasInternet” mostrar el promedio y suma total de la columna: Ingresos Netos por Cod_Producto.

transformacion9 = transformacion8.groupBy(F.col("Cod_Producto")).agg(F.round(F.avg("Ingresos_Netos"),4).alias("Ingreso_Neto_Promedio"), F.round(F.sum("Ingresos_Netos"),4).alias("Suma_Ingresos_Netos"))
transformacion9.display()
```
